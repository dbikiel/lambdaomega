# Understanding $\Lambda$-$\Omega$ systems

The general formulation of Lambda-Omega systems is:

$\frac{dx}{dt} = \Lambda(r)x - \Omega(r)y$

$\frac{dy}{dt} = \Omega(r)x + \Lambda(r)y$

with $r^2 = x^2 + y^2$

which upon cartesian to polar coordinates change is converted to:

$x = r\cos(\theta)$

$y = r\sin(\theta)$

$\frac{dr}{dt} = r\Lambda(r)$

$\frac{d\theta}{dt} = \Omega(r)$

Under this formulation, $\Lambda(\vec r) = 0$ are the circular limit cycles.

## Calculation of the nullclines

The nullclines of the system are the values for $\frac{dx}{dt} = 0$ and $\frac{dy}{dt} = 0$

which for the system above are:

$ay^3 + bxy^2 + (\omega + ax^2)y - (\lambda - bx^2)x = 0$ (x-nullcline)

$ax^3 − bx^2y + (\omega + ay^2)x + (\lambda − by^2)y = 0$ (y-nullcline)

## Solution

```python
class LambdaOmega:
    """
    Solves the system integrating the polar coordinates
    """    
    def __init__(self, l, o, a, b):
        self.l = l
        self.o = o
        self.a = a
        self.b = b
        
    def lambda_function(self, r):
        return self.l - self.b*r*r

    def omega_function(self, r):
        return self.o + self.a*r*r
    
    def df_dt(self, t, f0):
        r = f0[0]
        theta = f0[1]
    
        #dr/dt
        dr_dt = r*self.lambda_function(r)
    
        #dtheta/dt
        dtheta_dt = self.omega_function(r)
    
        return [dr_dt, dtheta_dt]
    
    def ode_solve(self, r_0, theta_0, tini, tfin, nsteps):

        #times to integrate
        ts = np.linspace(tini, tfin, nsteps)

        #initial conditions
        self.r_0 = r_0
        self.theta_0 = theta_0
        f0 = [r_0, theta_0]

        #Solve
        sol = solve_ivp(self.df_dt, t_span = [tini, tfin], t_eval = ts, y0 = f0)
        
        r_t = sol.y[0]
        theta_t = sol.y[1]
        
        self.t = sol.t
        self.x_t = r_t*np.cos(theta_t) 
        self.y_t = r_t*np.sin(theta_t)
        self.dx_dt = self.lambda_function(r_t)*self.x_t - self.omega_function(r_t)*self.y_t
        self.dy_dt = self.omega_function(r_t)*self.x_t + self.lambda_function(r_t)*self.y_t
        
        return [self.t, self.dx_dt, self.dy_dt, self.x_t, self.y_t]
        
    def nullclines(self, extend = 2, npoints = 100):
        
        def x_nullcline(y):
            return self.a*y**3 + self.b*x_vals*y**2 + (self.o + self.a*x_vals**2)*y - (self.l - self.b*x_vals**2)*x_vals

        def y_nullcline(x):
            return self.a*x**3 - self.b*x**2*y_vals + (self.o + self.a*y_vals**2)*x + (self.l - self.b*y_vals**2)*y_vals

        self.x_null = []
        self.y_null = []
        self.vals = np.linspace(-extend,extend, npoints)
        for i in self.vals:
            x_vals = i
            y_vals = i
            self.x_null.append(fsolve(x_nullcline,0))
            self.y_null.append(fsolve(y_nullcline,0))
            
        return [self.vals, self.x_null, self.y_null]    
```

```python
class LambdaOmega2:
    """
    Solves the system integrating the cartesian coordinates and add a term to include white noise
    """
    def __init__(self, l, o, a, b, D = 0):
        self.l = l
        self.o = o
        self.a = a
        self.b = b
        self.D = D
        
    def lambda_function(self, r):
        return self.l - self.b*r*r

    def omega_function(self, r):
        return self.o + self.a*r*r
    
    def noise(self, D = 0):
        return D*np.random.uniform(-1,1)
    
    def df_dt(self, t, f0):
        X0 = f0[0]
        Y0 = f0[1]
    
        r = np.sqrt(X0**2 + Y0**2)
        
        #dX/dt
        dX_dt = self.lambda_function(r)*X0 - self.omega_function(r)*Y0 + self.noise(self.D)
    
        #dY/dt
        dY_dt = self.omega_function(r)*X0 + self.lambda_function(r)*Y0
    
        return [dX_dt, dY_dt]
    
    def ode_solve(self, X_0, Y_0, tini, tfin, timestep):

        #times to integrate
        nstep = int((tfin - tini)/timestep)
        ts = np.linspace(tini, tfin, nstep, endpoint = True)

        #initial conditions
        self.X_0 = X_0
        self.Y_0 = Y_0
        f0 = [X_0, Y_0]

        #Solve
        sol = solve_ivp(self.df_dt, t_span = [tini, tfin], t_eval = ts, y0 = f0, method='RK45')

        self.t = sol.t

        self.x_t = sol.y[0]
        self.y_t = sol.y[1]
        r_t = np.sqrt(self.x_t**2 + self.y_t**2)
        self.dx_dt = self.lambda_function(r_t)*self.x_t - self.omega_function(r_t)*self.y_t
        self.dy_dt = self.omega_function(r_t)*self.x_t + self.lambda_function(r_t)*self.y_t
        
        return [self.t, self.dx_dt, self.dy_dt, self.x_t, self.y_t]
        
    def nullclines(self, extend = 2, npoints = 100):
        
        def x_nullcline(y):
            return self.a*y**3 + self.b*x_vals*y**2 + (self.o + self.a*x_vals**2)*y - (self.l - self.b*x_vals**2)*x_vals

        def y_nullcline(x):
            return self.a*x**3 - self.b*x**2*y_vals + (self.o + self.a*y_vals**2)*x + (self.l - self.b*y_vals**2)*y_vals

        self.x_null = []
        self.y_null = []
        self.vals = np.linspace(-extend,extend, npoints)
        for i in self.vals:
            x_vals = i
            y_vals = i
            self.x_null.append(fsolve(x_nullcline,0))
            self.y_null.append(fsolve(y_nullcline,0))
            
        return [self.vals, self.x_null, self.y_null] 
```

## Analysis of changing simultaneously $\lambda$ and b 

Using the new class to solve the ODEs in the cartesian coordinates, it is easy to add noise to the x(t) coordinate

```python
# Test lambda and b for three different cases mantaining the ratio
for i in [0.25, 1, 2]:
    #initialize a system (l, o, a, b, D = 0)
    lambda_omega = LambdaOmega2(i, 1, 1, i, 0.1)

    # Integrates the system (r_0, theta_0, tini, tfin, timestep)
    t, dx_dt, dy_dt, x_t, y_t = lambda_omega.ode_solve(0.005, 0, 0, 30, 0.01)

    # Calculates the nullclines up to extend
    vals, x_nullcline, y_nullcline = lambda_omega.nullclines(extend = 3)

    fig = plt.figure(figsize=(21,7))
    
    #Trajectory
    plt.subplot(1,2,1)
    plt.title('lambda = ' + str(i) + ' b = ' + str(i))
    plt.plot(t, x_t, label = 'x')
    plt.plot(t, y_t, label = 'y')
    plt.xlabel('time')
    plt.legend();

    #Nullclines
    plt.subplot(1,2,2)
    plt.title('lambda = ' + str(i) + ' b = ' + str(i))    
    plt.plot(x_t, y_t, 'b--', label = 'Trajectory')
    plt.plot(vals,x_nullcline, 'r', label = 'x-nullcline')
    plt.plot(y_nullcline, vals, 'g', label = 'y-nullcline');
    plt.axis('equal')
    plt.xlim(-2, 2)
    plt.ylim(-2, 2)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend();
```

### Determining lifetime of the transient by curve_fitting

Given that the shape of the evolution of R is sigmoidal and that the shape does not depends on the initial conditions of the system, we will explore a non-linear curve fitting to the trajectories.

```python
def sigmoid(x, x0, k):
    """
    Sigmoid function between 0 and 1, variable slope k and x0 midpoint position
    """
    return 1 / (1 + np.exp(-k*(x-x0)))

def sigmoid2(x, x0, k, h):
    """
    Sigmoid function between 0 and h, variable slope k and x0 midpoint position
    """
    return h / (1 + np.exp(-k*(x-x0)))


# Non-linear curve fitting using sigmoid function
def fit_r(func, x_data, y_data):
    
    if func == sigmoid:
        p0 = [np.median(x_data),1] # this is an mandatory initial guess
    elif func == sigmoid2:
        p0 = [np.median(x_data),1, np.max(y_data)] # this is an mandatory initial guess
    
    popt, _ = curve_fit(func, x_data, y_data, p0)
    
    return popt
```

```python
#initialize a system (l, o, a, b, D = 0)
lambda_omega = LambdaOmega2(0.25, 1, 1, 0.25, 0)

# Integrates the system (r_0, theta_0, tini, tfin, timestep)
t, dx_dt, dy_dt, x_t, y_t = lambda_omega.ode_solve(0.005, 0, 0, 30, 0.01)

# extract information to fit
xdata = t
ydata = x_t**2 + y_t**2


# Fit the trajectory
popt = fit_r(sigmoid, xdata, ydata)
print('Midpoint: ' + str(np.round(popt[0],3)) + ', Slope: ' + str(np.round(popt[1],3)))

fig, ax = plt.subplots(figsize=(14,7))
ax.plot(xdata, ydata, 'k', alpha = 0.7, label = 'Trajectory')
ax.plot(xdata, sigmoid(xdata, *popt), 'r', linestyle = '--', label = 'Fit')
ax.legend()
ax.set(xlabel = 'Time', ylabel = 'R');
```

## Estimating $\lambda$ and b from k

By varying $\lambda$ and b simultaneously and then extracting the slope of the fitted trajectory, we can obtain a relationship between the $\lambda$/b ratio and the slope

```python
lambdas = np.array([0.1, 0.25, 0.5, 1, 2, 3, 4, 5, 7.5, 10])
ks = []
for i in lambdas:
    lambda_omega = LambdaOmega2(i, 1, 1, i, 0)
    t, dx_dt, dy_dt, x_t, y_t = lambda_omega.ode_solve(0.005, 0, 0, 30, 0.01)
    
    xdata = t
    ydata = x_t**2 + y_t**2
    popt = fit_r(sigmoid, xdata, ydata)
    #print(i, popt[0], popt[1], popt[2], popt[3])
    print("lambda: {0: 5.1f}, Midpoint: {1: 8.2f}, Slope: {2: 8.2f}".format(i, popt[0], popt[1]))
    ks.append(popt[1])
ks = np.array(ks)
```

For example, fitting a linear regression

```python
ks_1 = sm.add_constant(ks)
mod = sm.OLS(lambdas, ks_1)
res = mod.fit()
print(res.summary())
```

```python
fig, ax = plt.subplots(figsize=(14,7))
ax.plot(ks, lambdas, '*', markersize = 10, label = 'parameter k from sigmoid fit for each trajectory');

ax.plot(ks, res.params[1]*ks + res.params[0], 'r', linestyle = '-', label = 'linear fit');

ax.set(xlabel = 'k', ylabel = 'lambda or b')
ax.text(15, 4, '$\lambda$ = k x ' + str(np.round(res.params[1],3)) + ' ' +  str(np.round(res.params[0],3)), size=24, ha='center', va='center');
#ax.legend();
```

# Connecting the empirical shape of r(t) with the equations

The analysis show that $x(t)^2 + y(t)^2$ (which equals to $r(t)^2$) has a very close shape to a sigmoidal function. Can we derive then for $r(t)^2$ has a sigmoidal function?

We know that:

$\frac{dr}{dt} = r\Lambda(r)$

Then we have to integrate:

$\int_{r_0}^{r} \frac{dr'}{r'\Lambda(r')} = \int_{0}^{t} dt'$

$\int_{r_0}^{r} \frac{dr'}{r'(\lambda - br'^2)} = \int_{0}^{t} dt'$


```python
from sympy import *
from sympy.plotting import plot
```

```python
x = Symbol('x')
l = Symbol('\lambda')
b = Symbol('b')
t = Symbol('t')
t0 = Symbol('t_0')
r0 = Symbol('r_0')
rt = Symbol('r_t')
```

The left side of the equation is:

```python
sol = integrate((x*(l - b*x**2))**-1,(x, r0, rt))
sol
```

The right hand is merely $t$, solving for $r(t)^2$

```python
solution = (solve(Eq(sol, t),rt)[1])**2
solution
```

The solution shows clearly the dependence of the evolution of $r^2$ with time depends on $e^{\lambda t}$

Replacing $\lambda / b$ by LC, it is possible to obtain:

$r(t)^2 = \frac{LC e^{2 \lambda t}}{ \frac{LC}{r_0^2} - 1  + e^{2 \lambda t}}$


Substituting $\lambda$, b and $r_0$ (which equals to $x_0^2 + y_0^2$):

```python
val = 0.25
p1 = plot(solution.subs({l:val, b:val, r0:0.005, t0:0}), (t, 0, 30))
```

For example, the solution shows that for cases where $\lambda = b$, the expression reduces to:

$r(t)^2 = \frac{e^{2 \lambda t}}{r_0^{-2} - 1  + e^{2 \lambda t}}$
